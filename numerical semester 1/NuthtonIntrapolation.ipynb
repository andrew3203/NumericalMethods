```python
import numpy as np
import matplotlib.pyplot  as plt
import math as m
```

```python
def Interpolation(x,u,a):
    N = len(x) - 1
    U = np.zeros((N+1,N+1))
    U[:, 0] = u
   
    for k in range(1,N+1):   
        for n in range(N + 1 - k):
            U[n,k] = k*(U[n,k-1]-U[n+1,k-1])/(x[n]-x[n+k])
    sm = 0.
    for n in range(N+1):
        mult = 1.
        for k in range(n):
            mult = mult*(a - x[k])   
        sm += U[0,n]/m.factorial(n)*mult
    return sm
```

```python
x = np.linspace(0, 2.6, 10)
u = np.sin(x)
fig, ax = plt.subplots(figsize=(8, 4))
a = 1
intr = Interpolation(x, u, a)
ax.plot(x, u, '-p')
ax.scatter([a], [intr], color='red')
```
